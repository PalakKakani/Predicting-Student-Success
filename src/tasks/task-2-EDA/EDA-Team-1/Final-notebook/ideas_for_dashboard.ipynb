```python
#  Loading the Libraries
import pandas as pd
import numpy as np
import chart_studio.plotly as py 
import plotly.express as px 
import cufflinks as cf
import plotly.graph_objects as go

from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected = True)
cf.go_offline()
```

```python
df = pd.read_csv('./Data/Final_feat.csv')
```

```python
df['avg_alc'].unique()
```

```python
fig = px.scatter(df, x="avg_alc", y="Final_Score", color="Subject")
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['age'], y=df['Final_Score'], mode='markers', name='Age vs Final Score'))
fig.update_layout(title='Scatter Plot: Age vs Final Score', xaxis_title='Age', yaxis_title='Final Score')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(x=df['school'], y=df['Final_Score'], name='Final Score vs School', boxpoints='all'))
fig.update_layout(title='Box Plot: Final Score vs School', xaxis_title='School', yaxis_title='Final Score')
fig.show()
```

### 3D Plots

```python
fig = px.violin(df, x='reason', y='Final_Score', color='sex', box=True, points='all',
                title='Violin Plot: Final Score vs Reason and Gender',
                labels={'reason': 'Reason', 'Final_Score': 'Final Score', 'sex': 'Gender'})
fig.show()
```

```python
# Parallel Coordinates Plot:

# The parallel coordinates plot allows you to compare the values of multiple numerical variables across different data points.
fig = px.parallel_coordinates(df, dimensions=['age', 'traveltime', 'studytime', 'Final_Score'],
                              color='Final_Score', color_continuous_scale='Viridis',
                              title='Parallel Coordinates: Age, Travel Time, Study Time, and Final Score')
fig.show()
```
