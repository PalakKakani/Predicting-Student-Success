```python
# to check if the dataset has any outliers we can use box plot.


cols = ['age', 'Medu', 'Fedu','traveltime', 'studytime','failures', 'famrel', 'freetime', 'goout', 'Dalc','Walc', 'health', 'absences', 'G1', 'G2', 'G3'],
      

for i in cols:
    sns.boxplot(df1[i], whis=1.5)
    plt.grid()
    plt.show()
```

```python
# to get the co relation value
df1.corr()
```

```python
# to visualize the corelation we are using heat map
correlation_matrix = df1.corr()
plt.figure(figsize=(35, 35))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

To get the target variable, we are going to create a new feature called final_score by using the scores of G1, G2 and G3.

As shown in the heat map the scores of G1, G2 and G3 has highest correlations so lets first see the correlation of G1, G2 and G3 seperately.

```python
# Creating a dataframe of G1,G2 and G3
G123 = df1[['G1', 'G2','G3']]
```

```python
# lets get the corelation vlaues
G123.corr()
```

```python
# lets visualize a heat map for the G123

plt.figure(figsize=(10,5))
sns.heatmap(G123.corr(), annot = True)
```

```python
# lets create a new feature called final_score by using the featurea G1, G2 and G3
# we have given weightage for the each features(for G1-30%, G2-30%, G3-40%)

df1['final_score'] = 0.3 * df1['G1'] + 0.3 * df1['G2'] + 0.4 * df1['G3']
```

```python
# Lets see the data
df1.head()
```

```python
# To drop G1, G2, and G3 columns
df1.drop(columns=['G1', 'G2', 'G3'], inplace=True)
```

```python
# Here we are encode binary categorical variables
binary_vars = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup','paid', 'activities',
               'nursery', 'higher', 'internet', 'romantic']

df1[binary_vars] = df1[binary_vars].replace({'GP': 0, 'MS': 1,'yes': 1, 'no': 0, 'F': 0, 'M': 1, 'U': 0, 'R': 1, 'LE3': 0, 'GT3': 1,
                                               'T': 0, 'A': 1})
```

```python
# to check whether the values are changed or not
df1.head()
```

```python
# Doing One-hot encode nominal categorical variables (Mjob, Fjob, reason, guardian)
df1 = pd.get_dummies(df1, columns=['Mjob', 'Fjob', 'reason', 'guardian'], drop_first=True)
```

```python
# to verify
df1.head()
```

```python
# to save the pre processed file into csv
df1.to_csv('pre_processed_student-mat.csv')
```
