```python
num_rows_df1 = df1.shape[0]
num_rows_df2 = df2.shape[0]
```

```python
for i in range(min(num_rows_df1, num_rows_df2)):
    row_df1 = df1.loc[i]
    row_df2 = df2.loc[i]
    if row_df1.equals(row_df2):
        print(f"Row {i+1} is the same in both datasets.")
    else:
        print(f"Row {i+1} is different in both datasets.")
```
