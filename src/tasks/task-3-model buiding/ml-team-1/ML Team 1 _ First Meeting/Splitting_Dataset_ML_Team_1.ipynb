```python
class cfg():
    df_path = "/Users/user/pythonpractice/Omdena/Predicting Student Sucess Omdena Ankara 2023/student-mat.csv"

    test_size = 0.3
```

```python
import pandas as pd

weight_g1 = 0.3
weight_g2 = 0.3
weight_g3 = 0.4

df = pd.read_csv(cfg.df_path, delimiter=";")

df.insert(0, "id", range(len(df)))

def calculate_final_score(df):
    df["Final_Score"] = (df['G1'] * weight_g1) + (df['G2'] * weight_g2) + (df['G3'] * weight_g3)
    return df

df = calculate_final_score(df)
```

```python
test = df.sample(frac=cfg.test_size, random_state=42)

train = df.drop(test.index)

train.to_csv("train-student-mat.csv")
test.to_csv("test-student-mat.csv")
```
